```python
import random
def play_game():
 
    print("WELCOME TO GUESS ME!")
 
    print("*" * 30)
 
    print("\n")
 
    print("I'm thinking of a number between 1 and 100")
 
    print("\n")
 
    print("If your guess is more than 10 away from my number, I'll tell you you're COLD")
 
    print("\n")
 
    print("If your guess is within 10 of my number, I'll tell you you're WARM")
 
    print("\n")
 
    print("If your guess is farther than your most recent guess, I'll say you're getting COLDER")
 
    print("\n")
 
    print("If your guess is closer than your most recent guess, I'll say you you're getting WARMER")
 
    print("\n")
 
    print("LET'S PLAY!")
 
    print("*" * 30)
 
 
 
    num = random.randint(1,100)
 
    guesses = [0]
 
   
 
    while True:
 
   
 
        #taking the right input
 
        while True:
 
            try:
 
                guess = int(input("I'm thinking of a number between 1 and 100.\n    What is your guess?"))
                
                # check for range 1-100    - CAIO          
                if 0 < guess < 100:
                    break
                else:
                    continue
 
            except ValueError:
 
                print("Oops, that wasn't a whole number. Please try again....")
 
        #guess = input("I'm thinking of a number between 1 and 100.\n What is your guess?")
 
   
 
        #if guess < 1 or guess > 100 or guess == float:
 
        #    print("OUT OF BOUNDS! Please try again A WHOLE NUMBER:")
 
        #    continue
 
       
 
        # here we can compare the player's guess to our number
 
        if guess == num:
 
            print("*" * 30)
 
            print(f'CONGRATULATIONS, YOU GUESSED IT IN ONLY {len(guesses)} GUESSES!')
 
            break
 
       
 
        # if guess is incorrect, add guess to the list
 
        guesses.append(guess)
 
   
 
        # when testing the first guess, guesses[-2]==0. which evaluates to False
 
        # and brings us down to the second section
 
   
 
        if guesses[-2]:
 
            if abs(num-guess) < abs(num-guesses[-2]):
 
                print("WARMER!")
 
            else:
 
                print("COLDER!")
 
   
 
        else:
 
            if abs(num-guess) <= 10:
 
                print("WARM!")
 
            else:
 
                print("COLD!")
 
   
 
        if len(guesses) == 11:
 
            print("*" * 30)
 
            print("Sorry, you have used all your 10 attempts! Please try again!")
 
            print("*" * 30)
 
            break
 
   
 
if __name__ == "__main__":
    while True:
        play_game()
        goAgain = input(f"Play Again? (Press Enter to continue, or q to quit):")
 
        # placing this outside in a while loop will make the game loop whenever the function end - CAIO
        if goAgain =='q':
            break
        else:
            continue
 
   
 
print("\n\n\n-------------------------------")
 
print('Thanks for Playing! Bye!')
```
